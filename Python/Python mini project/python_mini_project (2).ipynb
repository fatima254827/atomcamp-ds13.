```python
# 11. Write a function mission_success_rate(destination) that returns the percentage of success

def mission_success_rate(destination):  # define function mission success rate
    # filter missions for this destination
    missions = df[df["destination"] == destination]

    if len(missions) == 0:  # if there is no mission
        return "No missions found."
    total = len(missions)   # count total missions

    success_count = (missions["success"] == True).sum()   # count how many are successful (True)

    success_rate = (success_count / total) * 100   # calculate percentage
    return success_rate
# test data
print("Moon Success Rate:", mission_success_rate("Moon"), "%")
print("Mars Success Rate:", mission_success_rate("Mars"), "%")
print("ISS Success Rate:", mission_success_rate("ISS"), "%")
```

```python
# 12. Write a function longest_mission() that returns the mission name and duration of the longest mission.
def longest_mission():
    longest = df.loc[df["duration_days"].idxmax()]  # find the row with the maximum duration

    mission_name = longest["mission_name"] # extract mission name and duration
    duration = longest["duration_days"]

    return mission_name, duration

name, days = longest_mission()
print(f"The longest mission is '{name}' with a duration of {days} days.")
```

```python
# 13. Write a function missions_by_year(year) that returns all missions launched in a given year. Test with 2010
def missions_by_year(year):

    missions = df[df["year"] == year]   # filter missions for the year

    if missions.empty:   # condition if no mission found
        return "No missions found."

    return missions[["mission_id", "mission_name", "destination", "year"]]

print(missions_by_year(2010))  # check for 2010
```

```python
# 14.Write a function fuel_check(fuel_level) that returns "Critical" if fuel level is less than 20, "Moderate" if fuel level is equal to 20 or between 20 and 70, and "Safe" if fuel level is greater than or equal to 70. Apply this to the last 50 missions
# Function for fuel check
def fuel_check(fuel_level):
    if fuel_level < 20:
        return "Critical"
    elif 20 <= fuel_level < 70:
        return "Moderate"
    else:
        return "Safe"

# Apply to the last 50 missions
last_50 = df.tail(50)  # select last 50 rows
last_50["fuel_status"] = last_50["fuel_level"].apply(fuel_check)

# Show results
print(last_50[["mission_id", "mission_name", "fuel_level", "fuel_status"]])
```

```python
# 15. Write a nested loop to count the number of missions per commander for each destination. Print results like: Moon: Amelia Vega: 1 mission Zara Sheikh: 1 mission Mars: Liam Carter: 2 missions.
 # nested loops
destinations = df["destination"].unique()

for dest in destinations:  # outer loop for each destination
    commanders = df[df["destination"] == dest]["commander"].unique()

    for cmd in commanders:  # inner loop for each commander
        count = len(df[(df["destination"] == dest) & (df["commander"] == cmd)])
        print(f"{dest}: {cmd}: {count} mission{'s' if count > 1 else ''}")
```

```python
# 16.Write a try-except block to safely convert user input into an integer for selecting a mission ID. If invalid, print "Invalid input. Please enter a number."
try:
    mission_id = int(input("Enter a mission ID: "))  # user input
    print(f"You selected mission ID: {mission_id}")
except ValueError:                                # handling value error
    print("Invalid input. Please enter a number.")
```

```python
# 17. Paste your code for Q13 (missions_by_year) and modify it to handle the case when no missions exist for a year. Return "No missions found."
def missions_by_year(year):
    try:
        missions = df[df["year"] == year]         # filter missions for the year

        if missions.empty:   # if no missions exist
            return "No missions found."
        return missions[["mission_id", "mission_name", "destination", "year"]]
    except Exception as e:
        return f"Error: {e}"
print(missions_by_year(2050))
```

```python
# 18. Write a function get_mission(mission_id) that returns mission details if ID exists or raises a ValueError if mission_id is not found. Handle it with try-except
def get_mission(mission_id):
    try:
        mission = df[df["mission_id"] == mission_id] # find mission by mission_id
        if mission.empty:   # if no mission found
            raise ValueError("Mission ID not found.")
        return mission  # return mission details
    except ValueError as ve:
        return f"Error: {ve}"
    except Exception as e:  # any other unexpected error
        return f"Unexpected error: {e}"
print(get_mission(5))
print(get_mission(999))  # mission_ID doesnot exixt lets r this
```

```python
# 19.Ask user for a mission ID, and if input is not a number, print "Please enter digits only." If the mission itself does not exist at all, print "Mission not found.
try:
    user_input = input("Enter a mission ID: ")   # ask user input
    if not user_input.isdigit():     # check if input is digits only
        print("Please enter digits only.")
    else:
        mission_id = int(user_input)  # convert to integer
        mission = df[df["mission_id"] == mission_id]    # check if mission exists
        if mission.empty:
            print("Mission not found.")
        else:
            print("Mission details:")
            print(mission)

except Exception as e:
    print("Unexpected error:", e)
```

```python
# 20. Paste your code for Q19 and use finally to print "Mission lookup complete." after it is executed, no matter what happens.
try:
    user_input = input("Enter a mission ID: ")     # ask user input
    if not user_input.isdigit():     # check if input is digits only
        print("Please enter digits only.")
    else:
        mission_id = int(user_input)  # convert to integer
        mission = df[df["mission_id"] == mission_id]  # check if mission exists
        if mission.empty:
            print("Mission not found.")
        else:
            print("Mission details:")
            print(mission)
except Exception as e:             # incase error happende
    print("Unexpected error:", e)

finally:                               # it will run no matter what
    print("Mission lookup complete.")
```
