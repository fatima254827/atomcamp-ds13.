1. Largest Difference in List  
Given a list of integers, return the largest difference between any two elements, where the
larger element appears after the smaller one. If no such pair exists, return 0.  
Example:  
Input: [1, 2, 90, 10, 110]  
Output: 109  
Explanation: The largest difference is between 1 and 110.  

```python
def largest_difference(nums):
    max_diff = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            diff = nums[j] - nums[i]
            if diff > max_diff:
                max_diff = diff
    return max_diff

print(largest_difference([1, 2, 90, 10, 110]))
```

characters, return None.  
Example:  
Input: "aabcc"  
Output: 'b'  
Explanation: The first non-repeating character is 'b'.

3. Check If Strings Are Permutations  
Problem:  
Given two strings, write a function to check if one is a permutation of the other. Return True
if they are permutations, False otherwise.  
Example:  
Input: "abc", "bca"  
Output: True  
Explanation: Both strings contain the same characters in any order.  

```python
def are_permutations(s1, s2):
    if len(s1) != len(s2):
        return False
    return sorted(s1) == sorted(s2)

print(are_permutations("abc", "bca"))
print(are_permutations("abc", "abb"))
```

4. Find the Second Largest Unique Number  
Problem:  
Given a list of integers, return the second largest unique number. If no second largest unique
number exists, return None.  
Example:  
Input: [4, 5, 6, 6, 7]  
Output: 6  
Explanation: The second largest unique number is 6.

```python
def second_largest_unique(nums):
    uniq = list(set(nums))
    uniq.sort(reverse=True)
    if len(uniq) < 2:
        return None
    return uniq[1]
print( second_largest_unique([4, 5, 6, 6, 7]))
```

5. Count Distinct Pairs  
Problem:  
Given a list of integers and a target sum, count how many distinct pairs of integers in the list
sum up to the target.  
Example:  
Input: [1, 2, 3, 4, 3, 6], Target: 6  
Output: 2  
Explanation: The pairs (2, 4) and (1, 5) sum up to 6.  

```python
def count_distinct_pairs(nums, target):
    pairs = set()
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                a, b = min(nums[i], nums[j]), max(nums[i], nums[j])
                pairs.add((a, b))

    return len(pairs)
print( count_distinct_pairs([1, 2, 3, 4, 3, 6], 6))
```

6. Remove Duplicates from List  
Problem:  
Given a list of integers, remove all duplicates while maintaining the order of their first
occurrence.  
Example:  
Input: [1, 2, 2, 3, 4, 5, 1]  
Output: [1, 2, 3, 4, 5]  
Explanation: Duplicates are removed, but the order of elements is preserved.

```python
def remove_duplicates_preserve_order(nums):
    check = set()
    result = []
    for x in nums:
        if x not in check:
            check.add(x)
            result.append(x)
    return result

print(remove_duplicates_preserve_order([1,2,2,3,4,5,1]))
```

7. Subsets of a List  
Problem:  
Given a list of integers, return all possible subsets (the power set) of the list.  
Example:  
Input: [1, 2, 3]  
Output: [[], [1], [2], [3], [1, 2], [1, 3], [2, 3], [1, 2, 3]]
Explanation: The function returns all subsets of the list.

```python
def subsets(lst):
    result = [[]]
    for x in lst:
        new_subsets = []
        for subset in result:
            new_subsets.append(subset + [x])
        result += new_subsets
    return result
print(subsets([1, 2, 3]))
```

8. Longest Increasing Subsequence  
Problem:  
Given a list of integers, find the length of the longest subsequence of strictly increasing
elements.  
Example:  
Input: [10, 9, 2, 5, 3, 7, 101, 18]  
Output: 4  
Explanation: The longest increasing subsequence is [2, 3, 7, 101].

```python
def length_of_LIS(nums):
    if not nums:
        return 0
    n = len(nums)
    dp = [1] * n
    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
print(length_of_LIS([10,9,2,5,3,7,101,18]))
```

9. Maximum Product of Two Numbers  
Problem:  
Given a list of integers, return the maximum product of any two numbers in the list.  
Example:  
Input: [1, 10, -5, 1, 100]  
Output: 1000  
Explanation: The largest product comes from 10 * 100 = 1000.

```python
def max_product(nums):
    nums.sort()
    prod1 = nums[-1] * nums[-2]
    prod2 = nums[0] * nums[1]
    return max(prod1, prod2)

print(max_product([1, 10, -5, 1, 100]))
```

10. Frequency of Characters in String  
Problem:  
Given a string, return the character that appears most frequently. If there is a tie, return the
lexicographically smallest character.  
Example:  
Input: "aabbbc"  
Output: 'b'  
Explanation: The most frequent character is 'b'.


```python
def most_frequent_char(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    max_count = max(freq.values())
    candidates = [char for char in freq if freq[char] == max_count]
    return min(candidates)

print(most_frequent_char("aabbbc"))
```

11. Print Prime Numbers Less Than N  
Problem:  
Write a function that prints all prime numbers less than a given number n.  
Example:  
Input: 10  
Output: 2, 3, 5, 7

```python
def print_primes_less_than_n(n):
    for num in range(2, n):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=" ")
print_primes_less_than_n(10)
```

12. Sum of Variable-Length Arguments  
Problem:  
Write a function that takes a variable number of arguments (*args) and returns the sum of all
the numbers.  
Example:  
Input: (2, 3, 4)  
Output: 9  
Explanation: The sum of 2 + 3 + 4 is 9.  

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_numbers(2, 3, 4))
```

13. Find the Student with Highest Score  
Problem:  
Given a dictionary of students' names and their scores, write a function that returns the name of
the student with the highest score.  
Example:  
Input: {"Alice": 85, "Bob": 90, "Charlie": 92}  
Output: "Charlie"

```python
def top_student(scores):
    highest_score = -1
    best_student = None
    for name, score in scores.items():
        if score > highest_score:
            highest_score = score
            best_student = name
    return best_student
print(top_student({"Alice": 85, "Bob": 90, "Charlie": 92}))
```

14. Apply Lambda to List  
Problem:  
Write a function that takes a list of integers and a lambda function, and returns a new list where
the lambda function is applied to each element of the list.  
Example:  
Input: [1, 2, 3, 4], Lambda: lambda x: x ** 2  
Output: [1, 4, 9, 16]  
Explanation: Each number is squared by the lambda function.

```python
def apply_lambda(nums, func):
    result = []
    for n in nums:
        result.append(func(n))
    return result

print(apply_lambda([1, 2, 3, 4], lambda x: x ** 2))
```

15. Sum or Multiply Based on Keyword Argument  
Problem:  
Write a function that takes a variable-length list of numbers (*args) and an optional keyword
argument operation. If operation='multiply', return the product of all the numbers;
 otherwise, return their sum.  
Example:  
Input: (1, 2, 3, 4), operation='multiply'  
Output: 24  
Explanation: The product of the numbers 1 * 2 * 3 * 4 is 24.  

```python
def calculate(*args, operation="sum"):
    return sum(args) if operation != "multiply" else __import__("functools").reduce(lambda x, y: x*y, args)
print(calculate(1, 2, 3, 4, operation="multiply"))
```
